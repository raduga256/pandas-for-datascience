## Filtering Data


```python
# Filtering our records where Type 1 = 'Grass' and Type 2 = 'Poison'
new_df = df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison')]

new_df
```

```python
#Original Indxes are maintained on filtered data, so sometimes uhave to reset indexes
new_df.reset_index(drop=True, inplace = True)

new_df
```

## Conditional Changes

```python
#filtering and then changing a certain column values e.g 'Fire' to Flamer
df.loc[df['Type 1']== 'Fire', 'Type 1'] = 'Flamer'

#reversing the change for column 'Type 1'
#df.loc[df['Type 1']== 'Flamer', 'Type 1'] = 'Fire'

df
```

```python
# conditing updating multiple columns at once using lists
df.loc[df['Total'] > 500, ['Speed', 'Legendary']] = [50, 'Test 2']
df2 = df[df['Legendary']== 'Test 2']

df2.head(4)
```

## Aggregate Statistics(groupby)

```python
# Lets rollback to the last checkpoint of our df that doesn't contain recent  changes
df = pd.read_csv('modified.csv')
df
```

```python
# averages values per groups in column
df.groupby(['Type 1']).mean().sort_values('Defense', ascending =False)
```

## Working with very Large amounts of data
May be your dealing with a large file say 20GB, you can not load it in at once, pandas allows us to load very large data set in chunks

```python
#chunksize stands for number of rows

#create a new df with same column names as the original large file, then add the read in
#chunks sequentially 

new_df = pd.DataFrame(columns=df.columns)

for df in pd.read_csv('modified.csv', chunksize= 5):
    #print('CHUNK DF')
    #print(df)
    results = df.groupby(['Type 1']).count()  #perform a statistic and save 
    
    #save the summarised data of interest and will be a lot smaller than the original data frame
    new_df = pd.concat([new_df, results])
    
    
    
```
